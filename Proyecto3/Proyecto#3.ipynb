### Extraccion de datos

```python
import pandas as pd #Importar la libreria pandas con el alias pd
data = pd.read_excel("Proyecto#3.xlsx") #Leer datos de un excel
```

### Analisis de datos

```python
data["tienda"].unique() #Obtener todos los valores unicos de una columna
```

```python
data["tienda"].value_counts() #Contar la cantidad de veces que se repite cada valor unico de una columna en orden descendente
```

```python
data.groupby("tienda")["precio"].sum() #Agrupar las tiendas con la suma de sus precios en orden ascendente
```

```python
data.groupby(["tienda", "forma_pago"])["precio"].sum() #Agrupar las varias columnas con la suma de sus precios en orden ascendente
```

```python
data.groupby("forma_pago")["precio"].sum().to_frame() #Mostrar los resultados en formato de DataFrame
```

```python
datos_agrupados = data.groupby("forma_pago")["precio"].sum().to_frame() #Asignar tabla agrupada a una variable
```

### Guardado de datos

```python
datos_agrupados.to_excel("datos_agrupados.xlsx") #Exportar la tabla agrupada a un nuevo archivo de excel
```

### Visualizacion de datos

```python
import plotly.express as px #Importar la libreria plotly con el alias pl
grafico =px.histogram(data,
              x="tienda",
              text_auto=True,
              title="Histograma de Formas de Pago",
              color="forma_pago") #Generar un histograma con parametros
```

```python
grafico.show() #Mostrar el grafico generado
```

```python
grafico.write_html("grafico.html") #Exportar el grafico a un archivo html
```
